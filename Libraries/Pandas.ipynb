Pivot Tables

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values='Sales',index = 'Region',columns='Product')#mean
```

```python
pd.pivot_table(df,values='Sales',index = 'Region',columns='Product',aggfunc='median')
```

```python
pd.pivot_table(df,values=['Sales','Units'],index = 'Region',columns='Product')
```

```python
#Cross Table
pd.crosstab(df['Region'],df['Product'])
```

```python
#Feature Extraction
def extract_episode(txt):
    start,end = 0,0
    for i in range(len(txt)):
        if txt[i] == "(":
            start = i+1
        if txt[i] == ")":
            end = i
    return txt[start:end]
df['Episode']=pd.DataFrame(df['Title'].apply(extract_episode))
df.head()
```

```python
df['Episode'] = df['Episode'].str.replace("eps","")
df['Episode'] = df['Episode'].astype(int)
df.head()
```

```python
df.head()
```

```python
df.loc[0]['Title']
```

```python
df.head()
```

```python
#Feature Extraction
def extract_episode(txt):
    start = 0
    for i in range(len(txt)):
        if txt[i] == ")":
            start = i+1
    return txt[start:start+19]
df['Time-Stamp'] = pd.DataFrame(df['Title'].apply(extract_episode))
df.head()
```

```python
import pandas as pd

def months_between(period):
    start, end = period.split(" - ")
    start = pd.to_datetime(start, format="%b %Y")
    end = pd.to_datetime(end, format="%b %Y")
    return (end.year - start.year) * 12 + (end.month - start.month)

df["Months"] = df["Time-Stamp"].apply(months_between)

df.head()
```
